# Importing File

```
import pandas as pd
df = pd.read_csv('diabetes.csv')
```

```
df
```

# Identifying features and labels

```
x = df.iloc[:, :-1] # last column is label so excluding it from x
y = pd.DataFrame(df.iloc[:,-1])
```

# Normalizing Input Data

```
mean = x.mean(axis=0)
x = x - mean
std = x.std(axis=0)
x /= std
```

# Splitting in Training & Test data

```
x_train, x_test = x.iloc[:int(len(x)*0.7)], x.iloc[:int(len(x)*0.7)] #70-30 split
y_train, y_test = y.iloc[:int(len(y)*0.7)], y.iloc[:int(len(y)*0.7)]
```

```
display(x_train)
display(x_test)
```

```
from keras.models import Sequential
from keras.layers import Dense, Dropout

model = Sequential()
model.add(Dense(93, input_shape = (x.shape[1],), activation = 'tanh'))
model.add(Dense(75, activation = 'tanh'))
model.add(Dense(1, activation = 'sigmoid'))

model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
model.summary()
```

```
model.fit(x_train, y_train, validation_data= (x_test, y_test), epochs = 100)
```
